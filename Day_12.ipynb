```python
from gensim.models import doc2vec
from gensim.models.doc2vec import TaggedDocument,Doc2Vec
```

```python
with open("J. K. Rowling - Harry Potter 1 - Sorcerer's Stone.txt","r")as f:
    text = f.read()
    
```

```python
import re

pattern = r"[^A-Za-z. ]"

processed_text = re.sub(pattern,"",text)
```

```python
processed_text
```

```python
sents = processed_text.split(".")
```

```python
sents[0]
```

```python
from nltk.tokenize import word_tokenize
```

```python
for id,sent in enumerate(sents):
    TaggedDocument(words=word_tokenize(sent),tags=[str(id)])
    
```

```python
import nltk
nltk.download("punkt")
```

```python
tagged_data = [TaggedDocument(words=word_tokenize(sent),tags=[str(id)]) for id,sent in enumerate(sents)]
```

```python
tagged_data[5]
```

```python
model = Doc2Vec(vector_size=100)
```

```python
model.build_vocab(tagged_data)
```

```python
len(model.wv.key_to_index)
```

```python
model.corpus_count
```

```python
model.train(tagged_data,total_examples=model.corpus_count,epochs=20)
```
