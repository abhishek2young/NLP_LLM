```python
with open("J. K. Rowling - Harry Potter 1 - Sorcerer's Stone.txt","r")as f:
    text = f.read()
    
```

```python
text
```

```python
import re
```

```python
pattern = r"[^A-Za-z. ]"
processed_text = re.sub(pattern,"",text)
```

```python
processed_text
```

```python
text_list = processed_text.split(".")
```

```python
text_list
```

```python
sentences = [sent.split() for sent in text_list]
```

```python
sentences
```

```python
sentences[0]
```

```python
from gensim.models import FastText
```

```python
model = FastText(vector_size=100,window=5)
```

```python
model.build_vocab(sentences)
```

```python
model.train(corpus_iterable=sentences,total_examples=len(sentences),epochs=10)
```

```python
model.wv["harry"]
```
