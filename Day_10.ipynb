```python
import gensim
```

```python
with open("J. K. Rowling - Harry Potter 1 - Sorcerer's Stone.txt","r")as f:
    text = f.read()
```

```python
text
```

```python
import re
```

```python
pattern = r"[^a-zA-Z. ]"
processed_text = re.sub(pattern,"",text)
```

```python
processed_text
```

```python
text_list = processed_text.split(".")
```

```python
text_list
```

```python
sentences_split = [sent.split() for sent in text_list]
```

```python
sentences_split[1000]
```

```python
len(sentences_split)
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences=sentences_split,sg=1,vector_size=100,window=5,min_count=1)
```

```python
print(model.wv.index_to_key)
```

```python
len(model.wv.index_to_key)
```

```python
len(model.wv["Harry"])
```

```python
model.wv["Hermione"]
```

```python
model.wv.most_similar("Hermione")
```

```python
model.save("Harry_Potter_word2vec.model")
```
